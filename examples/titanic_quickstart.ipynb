# Zedda Quickstart (Titanic Dataset)
This notebook demonstrates how to use Zedda to instantly profile, clean, and analyze datasets.

```
import zedda as zd
```

## 1. Full Dataset Profile
Run a blazing fast EDA profile on the titanic dataset.

```
zd.profile(
    "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
)
```

## 2. Smart Warnings
Get actionable data quality warnings.

```
zd.warnings(
    "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
)
```

## 3. Auto-Clean
Fix the missing values and drop ID columns automatically.

```
# apply=True directly returns a cleaned pandas DataFrame!
clean_df = zd.clean(
    "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv",
    apply=True,
)
clean_df.head()
```
